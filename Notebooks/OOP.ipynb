### 3) Add a method that will allow the Rocket to move up.

```python
# Building a class with some attributes
class Rocket():


  # use special method in python to initialize the attributes from this class
  def __init__(self):
    self.x = 0
    self.y = 0

  # Add a method that will make the Rocket Move Up(i.e. Y coordinate) by 1 Unit whenever it is called.

  def move_up(self):
    self.y += 1
```

```python
# Make a rocket named orion.
orion = Rocket()

print(f'Initial Possition of Orion :- {(orion.x, orion.y)} ')

# Move orion up by one unit.
orion.move_up()
print(f'Updated Possition of Orion, after it had move one unit up :- {(orion.x, orion.y)} ')

# Move orion up by one more unit.
orion.move_up()
print(f'Updated Possition of Orion after it had moved one more unit up :- {(orion.x, orion.y)} ')
```

### 6) Add another method to the Rocket class that calculates the distance between 2 Rockets.

```python
# Building a class with some attributes
class Rocket():


  # use special method in python to initialize the attributes from this class
  def __init__(self,x=0,y=0):
    self.x = x
    self.y = y



  # Add a method that will make the Rocket Move in either Up, Down, Left and Right with whatever units

  def move(self, x_incr =0, y_incr=1):
    self.x += x_incr
    self.y += y_incr


  # Calculate Eucledian distance between two rockets
  def get_distance(self, other_rocket):
    x_diff = self.x - other_rocket.x
    y_diff = self.y - other_rocket.y

    distance = (x_diff**2 + y_diff**2)**0.5
    return round(distance ,3)
```

```python
rocket4= Rocket(-10,0)
rocket5= Rocket(0,0)
print(f'{rocket4.x, rocket4.y}')
print(f'{rocket5.x, rocket5.y}')
```

```python
# Distance of Rocket 4 wrt Rocket 5
distance_4to5 = rocket5.get_distance(rocket4)
print(f'The distance between rocket 4 wrt rocket 5 is:- {distance_4to5} (3 decimal pts)')
```

```python
# Move rocket 4 to (-15,10)
rocket4.move(-5,10)

# Move rocket 5 to (0,15)
rocket5.move(0,15)


# Get the new locations of rocket 4 and rocket 5
print(f'{rocket4.x, rocket4.y}')
print(f'{rocket5.x, rocket5.y}')

```

```python
# New Distance of Rocket 4 wrt Rocket 5
distance_4to5 = rocket5.get_distance(rocket4)
print(f'The distance between rocket 4 wrt rocket 5 is:- {distance_4to5} (3 decimal pts)')
```
